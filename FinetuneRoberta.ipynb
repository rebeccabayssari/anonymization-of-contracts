```python
from datasets import Dataset
import pandas as pd

# Define the label mapping
label_list = ["O", "B-MISC", "I-MISC", "B-PER", "I-PER", "B-ORG", "I-ORG", "B-LOC", "I-LOC"]
label2id = {label: i for i, label in enumerate(label_list)}

# Read the train.txt file
def load_bio_data(file_path):
    sentences = []
    current_sentence = []
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip() #Reads the file line by line.
#Removes unnecessary spaces using .strip()
#If the line is NOT empty, it contains a word-label pair ‚Üí process it.
#If the line is empty, it indicates the end of a sentence
            if line:
                # Splits the line into a word and its label
                word, label = line.rsplit(" ", 1)
                current_sentence.append((word, label))
            else:
                if current_sentence:
                    sentences.append(current_sentence)
                    current_sentence = []
    #Handle Last Sentence
    if current_sentence:
        sentences.append(current_sentence)
    return sentences

# Convert to dataset format
def create_dataset(sentences):
    words = [[word for word, label in sent] for sent in sentences]
    labels = [[label2id[label] for word, label in sent] for sent in sentences]  # Convert to label IDs
    
    df = pd.DataFrame({"tokens": words, "ner_tags": labels})
    return Dataset.from_pandas(df)

# Load and convert the dataset
train_sentences = load_bio_data("train.txt")
train_dataset = create_dataset(train_sentences)

# Save the dataset
train_dataset.save_to_disk("ner_dataset")
print("‚úÖ Dataset saved successfully!")
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, TrainingArguments, Trainer
from transformers import DataCollatorForTokenClassification

model_name = "Alizee/xlm-roberta-large-finetuned-wikiner-fr"
tokenizer = AutoTokenizer.from_pretrained(model_name)


label_list = ["O", "B-MISC", "I-MISC", "B-PER", "I-PER", "B-ORG", "I-ORG", "B-LOC", "I-LOC"]
label2id = {label: i for i, label in enumerate(label_list)}
id2label = {i: label for label, i in label2id.items()}

# Load model with correct num_labels and ignore mismatched layer sizes
model = AutoModelForTokenClassification.from_pretrained(
    model_name, 
    num_labels=len(label_list),  # 9 labels instead of 5
    ignore_mismatched_sizes=True  # Fixes size mismatch
)

# Assign label mappings
model.config.id2label = id2label
model.config.label2id = label2id

# Tokenize dataset
def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(
        examples["tokens"], 
        truncation=True, #Cuts off long sequences exceeding the model's limit
        padding="longest",  # Ensures all sequences have the same length.
        is_split_into_words=True
    )
    #Align Labels with Tokenized Words
    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                label_ids.append(-100)  # Ignore padding tokens
            else:
                if label[word_idx] in label2id:
                    label_ids.append(label2id[label[word_idx]])  
                else:
                    label_ids.append(-100)  # Ignore unknown labels
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs

# Apply tokenization
tokenized_train = train_dataset.map(tokenize_and_align_labels, batched=True)

# Data collator for padding: It automatically pads sequences to the longest length in the batch during training.
data_collator = DataCollatorForTokenClassification(tokenizer)
```

```python
from transformers import TrainingArguments, Trainer

# Tokenize the entire dataset (train + validation)
tokenized_datasets = dataset.map(tokenize_and_align_labels, batched=True)

# Remove unused columns (only keep "input_ids", "attention_mask", "labels")
tokenized_datasets = tokenized_datasets.remove_columns(["tokens", "ner_tags"])

# Define training arguments
training_args = TrainingArguments(
    output_dir="./ner_results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    save_total_limit=2,
    load_best_model_at_end=True,  # Ensures best model is restored
    metric_for_best_model="eval_loss",  # Saves based on validation loss
    greater_is_better=False,  # Lower loss is better
    remove_unused_columns=False  # Important fix for Trainer
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],  # Use tokenized datasets
    eval_dataset=tokenized_datasets["validation"],  # Use tokenized datasets
    tokenizer=tokenizer,
    data_collator=data_collator
)

# Start training
trainer.train()

# Save the fine-tuned model and tokenizer
model.save_pretrained("./fine_tuned_xlm_roberta")
tokenizer.save_pretrained("./fine_tuned_xlm_roberta")

print("‚úÖ Model fine-tuned and saved successfully!")
```

```python
import spacy
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline

# Load SpaCy model for sentence segmentation
nlp = spacy.load("fr_core_news_sm")

def split_text_into_sentence_chunks(text, sentences_per_chunk=2):
    
    doc = nlp(text)
    sentences = [sent.text for sent in doc.sents]
    return [" ".join(sentences[i:i + sentences_per_chunk]) for i in range(0, len(sentences), sentences_per_chunk)]
#Combine sentences into chunks of sentences_per_chunk size

def group_entities(results, allowed_types={"ORG", "LOC", "PER"}):
    """Groups B- and I- tokens into complete entities."""
    grouped_entities = []
    current_entity = None

    for result in results:
        entity_type = result['entity_group']  
        word = result['word']
        start = result['start']
        end = result['end']
        #Skip entities not in the allowed_types list:
        if entity_type not in allowed_types:
            if current_entity:
                grouped_entities.append(current_entity)
                current_entity = None
            continue
        #If the current entity is the same type, append the new word.
        #Update end position for the extended entity  
        if current_entity and entity_type == current_entity["type"]:
            current_entity["text"] += " " + word
            current_entity["end"] = end
        #If a new entity is detected, save the previous entity and start a new one
        else:
            if current_entity:
                grouped_entities.append(current_entity)
            current_entity = {"text": word, "type": entity_type, "start": start, "end": end}
    #Add the last entity after looping through all words:
    if current_entity:
        grouped_entities.append(current_entity)

    return grouped_entities

# Load the fine-tuned model
model_path = "./fine_tuned_xlm_roberta"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForTokenClassification.from_pretrained(model_path)

# Create an NER pipeline
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer, aggregation_strategy="simple")  

# Load the test contract
#file_path = r"C:\Users\hp\Downloads\clever contact\clevercontracts\Contracts\2MAINS\test_contract.txt"
#file_path = "cleanedcontract2.txt"
#file_path = r"C:\Users\hp\Downloads\clever contact\cleanedcontract3.txt"
file_path = r"C:\Users\hp\Downloads\clever contact\cleanedcontract4.txt"



with open(file_path, "r", encoding="utf-8") as f:
    contract_text = f.read()

# Apply sentence chunking
sentence_chunks = split_text_into_sentence_chunks(contract_text, sentences_per_chunk=2)

# Apply the NER model and group results
all_grouped_results = []
for i, chunk in enumerate(sentence_chunks):
    print(f"\nChunk {i + 1}:\n{chunk}\n")
    results = ner_pipeline(chunk)
    grouped_results = group_entities(results, allowed_types={"ORG", "LOC", "PER"})  # Customize as needed
    all_grouped_results.extend(grouped_results)

# Display detected entities
print(" R√©sultat de la detection :")
for entity in all_grouped_results:
    print(f"Entity: {entity['text']} | Type: {entity['type']} | Start: {entity['start']} | End: {entity['end']}")




```
